## Stratégie d'imputation:<br>
### 1. La colonne cabin contient 1014 valeurs manquante d'un totale de 1309 valeurs ==> supprimer la colonne

```python
tdata = tdata.drop(['cabin'],axis=1)
```

```python
tdata.head()
```

### 2. Remplacer les valeurs manquantes de la colonne embarked par la valeur la plus redendante
* la méthode fillna() remplace les valeurs NaN par ce q'on lui demande*

```python
tdata['embarked'].isnull().sum()
```

```python
tdata.embarked.value_counts()
# déterminer et afficher le nombre d'observations par modalité
```

```python
tdata['embarked'] = tdata['embarked'].fillna(2)
tdata.head()
```

```python
tdata['embarked'].isnull().sum()
```

### 3. Remplacer les valeurs manquantes de la colonne fare par la valeur moyenne

```python
tdata['fare'].isnull().sum()
# afficher le nombre de valeurs manquantes
```

```python
tdata.fare.mean()
# afficher la valeur moyenne de la variable fare
```

```python
tdata['fare'] = tdata['fare'].fillna(tdata.fare.mean())
# la méthode fillna permet de remplacer les valeurs manquantes par la valeur moyenne (ou par la médiane )
```

```python
tdata['fare'].isnull().sum()
# vérifier que le traitement a été fait avec succés
```

### 4. Remplacer les valeurs manquantes de la colonne age par la valeur médiane

```python
NaN_indexes = tdata['age'][tdata['age'].isnull()].index
# affichage des index dont l'age est NaN
```

```python
tdata.age.median()
```

```python
tdata['age'] = tdata['age'].fillna(tdata.age.median())
```

```python
tdata.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
print(X_train.shape)
```

```python
KNN=KNeighborsClassifier(n_neighbors=5)
KNN.fit(X_train,y_train)
```

```python
y_pred = KNN.predict(X_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Évaluer les performances
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Affichage des résultats avec formatage
print(f"Exactitude du modèle : {accuracy:.2f}")
print("\nMatrice de confusion :\n", conf_matrix)
print("\nRapport de classification :\n", class_report)

# Visualisation de la matrice de confusion
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non Survécu', 'Survécu'],
            yticklabels=['Non Survécu', 'Survécu'])
plt.ylabel('Vérité Terrain')
plt.xlabel('Prédictions')
plt.title('Matrice de Confusion')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Définir les paramètres à tester
param_grid = {
    'n_neighbors': [1, 3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}

# Créer le classificateur KNN
KNN = KNeighborsClassifier()

# Configurer la recherche par grille
grid_search = GridSearchCV(estimator=KNN, param_grid=param_grid,
                           scoring='accuracy', cv=5, verbose=1)

# Entraîner le modèle avec recherche par grille
grid_search.fit(X_train, y_train)

# Meilleurs paramètres trouvés
best_params = grid_search.best_params_
print("Meilleurs paramètres : ", best_params)

# Meilleur modèle
best_model = grid_search.best_estimator_

# Faire des prédictions sur l'ensemble de test
y_pred = best_model.predict(X_test)

# Évaluer les performances
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Affichage des résultats
print(f"Exactitude du modèle : {accuracy:.2f}")
print("\nMatrice de confusion :\n", conf_matrix)
print("\nRapport de classification :\n", class_report)

# Visualisation de la matrice de confusion
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non Survécu', 'Survécu'],
            yticklabels=['Non Survécu', 'Survécu'])
plt.ylabel('Vérité Terrain')
plt.xlabel('Prédictions')
plt.title('Matrice de Confusion')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Créer le classificateur d'arbre de décision
dtree = DecisionTreeClassifier(random_state=42)

# Entraîner le modèle
dtree.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = dtree.predict(X_test)

# Évaluer les performances
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Affichage des résultats
print(f"Exactitude du modèle : {accuracy:.2f}")
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non Survécu', 'Survécu'],
            yticklabels=['Non Survécu', 'Survécu'])
plt.ylabel('Vérité Terrain')
plt.xlabel('Prédictions')
plt.title('Matrice de Confusion')
plt.show()
print("\nRapport de classification :\n", class_report)

# Visualiser l'arbre de décision avec une distance accrue entre les nœuds
plt.figure(figsize=(15, 10))
plot_tree(dtree, feature_names=X.columns, class_names=['Non Survécu', 'Survécu'],
          filled=True, rounded=True , proportion=True,
          node_ids=True, impurity=False)

plt.title("Arbre de Décision", fontsize=16)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Créer le classificateur de forêt aléatoire
rf_model = RandomForestClassifier(random_state=42)

# Entraîner le modèle
rf_model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = rf_model.predict(X_test)

# Évaluer les performances
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Affichage des résultats
print(f"Exactitude du modèle : {accuracy:.2f}")
plt.figure(figsize=(15, 10))
plot_tree(dtree, feature_names=X.columns, class_names=['Non Survécu', 'Survécu'],
          filled=True, rounded=True , proportion=True,
          node_ids=True, impurity=False)

plt.title("Arbre de Décision", fontsize=16)
plt.show()
print("\nRapport de classification :\n", class_report)

# Visualiser l'importance des caractéristiques
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Importance des caractéristiques")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=45)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Obtenir les probabilités de prédiction pour la classe positive
y_scores = best_model.predict_proba(X_test)[:, 1]  # Classe positive

# Calculer les valeurs de la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_scores)

# Calculer l'aire sous la courbe (AUC)
roc_auc = auc(fpr, tpr)

# Visualiser la courbe ROC
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonale
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('Taux de Faux Positifs (FPR)')
plt.ylabel('Taux de Vrais Positifs (TPR)')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
